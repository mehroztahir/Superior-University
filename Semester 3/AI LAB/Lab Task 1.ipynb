```python
import random

HANGMAN_PICS = ['''
  +---+
  |   |
      |
      |
      |
      |
========''', '''
  +---+
  |   |
  O   |
      |
      |
      |
========''', '''
  +---+
  |   |
  O   |
  |   |
      |
      |
========''', '''
  +---+
  |   |
  O   |
 /|   |
      |
      |
========''', '''
  +---+
  |   |
  O   |
 /|\  |
      |
      |
========''', '''
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
========''', '''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
========''']

words = ["baloon", "camel", "mehroz", "computer", "chair", "house", "plant"]
chosen_word = random.choice(words)
blank = ""
for letter in chosen_word:
    blank += "_"
blank_list = list(blank)

update_display = 0
guessed_letters = ""

def making_a_guess():
    global update_display
    global guessed_letters
    guess = input(f"Welcome to hangman.\n{''.join(blank_list)}\nMake a guess? ")
    if len(guess) != 1:
        print("Please enter a single letter.")
    elif guess in guessed_letters:
        print("You already guessed that letter, choose again.")
    elif guess not in chosen_word:
        print("That letter is not in the word.")
        update_display += 1
    else:
        print("Good job, that letter is in the word!")
        index = 0
        for letter in chosen_word:
            if letter == guess:
                blank_list[index] = guess
            index += 1
    guessed_letters += guess

while update_display < 6:
    making_a_guess()
    print(HANGMAN_PICS[update_display])
    print(''.join(blank_list))
    if '_' not in blank_list:
        print("Congratulations, you won!")
        break
else:
    print("Game over, the word was " + chosen_word)
```
